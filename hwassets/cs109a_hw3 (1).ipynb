# Forecasting Bike Sharing Usage

In this homework, we will focus on multiple linear regression and will explore techniques for subset selection. The specific task is to build a regression model for a bike share system that can predict the total number of bike rentals in a given day, based on attributes about the day. Such a demand forecasting model would be useful in planning the number of bikes that need to be available in the system on any given day, and also in monitoring traffic in the city. The data for this problem was collected from the Capital Bikeshare program in Washington D.C. over two years.   

The data set is provided in the files `Bikeshare_train.csv` and `Bikeshare_test.csv`, as separate training and test sets. Each row in these files contains 10 attributes describing a day and its weather:
- season (1 = spring, 2 = summer, 3 = fall, 4 = winter)
- month (1 through 12, with 1 denoting Jan)
- holiday (1 = the day is a holiday, 0 = otherwise)
- day_of_week (0 through 6, with 0 denoting Sunday)
- workingday (1 = the day is neither a holiday or weekend, 0 = otherwise)
- weather 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
- temp (temperature in Celsius)
- atemp (apparent temperature, or relative outdoor temperature, in Celsius)
- humidity (relative humidity)
- windspeed (wind speed)

and the last column 'count' contains the response variable, i.e. total number of bike rentals on the day.

We next require you to pre-process the categorical and numerical attributes in the data set:

- Notice that this data set contains categorical attributes with two or more categories. **Why can't they be directly used as predictors?** Convert these categorical attributes into multiple binary attributes using one-hot encoding: in the place of every categorical attribute $x_j$ that has categories $1, \ldots, K_j$, introduce $K_j-1$ binary predictors $x_{j1}, \ldots, x_{j,K_j-1}$ where $x_{jk}$ is 1 whenever $x_j = k$ and 0 otherwise. 
** Why is it okay to not have a binary column for the $K_j$-th category? **

- Since the attributes are in different scales, it is a good practice to standardize the continuous predictors, i.e. to scale each continuous predictor to have zero mean and a standard deviation of 1. This can be done by applying the following transform to each continuous-valued predictor $j$:
$\hat{x}_{ij} = (x_{ij} - \bar{x}_j) / s_j$, where $\bar{x}_j$ and $s_j$ are the sample mean and sample standard deviation (SD) of predictor $j$ in the training set. We emphasize that the mean and SD values used for standardization must be estimated using only the training set observations, while the transform is applied to both the training and test sets. 
** Why shouldn't we include the test set observations in computing the mean and SD? **

**Note:** We shall use the term "attribute" to refer to a categorical column in the data set, and the term "predictor" to refer to the individual binary columns resulting out of one-hot encoding.

### Answers
**Why can't they be directly used as predictors? **
Because categorical attributes are not numeric, they can not be put into computation. The categorical attributes with more than two categories cannot be directly used as predictors since the variance in the distances between any instances, which isn't really meaningful, could shadow the correlation between the response variable and these categorical attributes.

** Why is it okay to not have a binary column for the $K_j$-th category? **
Because it can be represented by other columns (in the const term). We can infer the category from the values of the first $K_j-1$ dummy variables (the case where the first $K_j-1$ dummy variables are all equal to zero corresponds to the $K_j$-th category). It is more economical to use $K_j-1$ variables.

** Why shouldn't we include the test set observations in computing the mean and SD? **
We are not supposed to "observe" the test dataset samples when we train the model.
Because we are normalize and train according the training data set. In reality, we never know what our test data would look like so we cant't include the mean and SD of test data.

```python
# Pre-process
data_train = pd.read_csv('data/Bikeshare_train.csv')
data_test = pd.read_csv('data/Bikeshare_test.csv')
data_train = data_train.drop('Unnamed: 0', 1)
data_test = data_test.drop('Unnamed: 0', 1)

data_train = pd.get_dummies(data_train, columns=['season', 'month', 'day_of_week', 'weather'])
data_train = data_train.drop(['season_4.0', 'month_12.0', 'day_of_week_6.0', 'weather_3.0'], 1)
data_test = pd.get_dummies(data_test, columns=['season', 'month', 'day_of_week', 'weather'])
data_test = data_test.drop(['season_4.0', 'month_12.0', 'day_of_week_6.0', 'weather_3.0'], 1)

cp = ['temp', 'atemp', 'humidity', 'windspeed']
cp_means = [data_train[v].mean() for v in cp]
cp_std = [data_train[v].std() for v in cp]
for i, v in enumerate(cp):
    data_train[v] = (data_train[v] - cp_means[i]) / cp_std[i]
    data_test[v] = (data_test[v] - cp_means[i]) / cp_std[i]

col_names = list(data_train.columns[7:]) + list(data_train.columns[:7])
data_train = data_train[col_names]
data_test = data_test[col_names]
data_train.describe()
```

## Part (d): Subset Selection

Apply either one of the following subset selection methods discussed in class to choose a minimal subset of predictors that are related to the response variable:
- Step-wise forward selection
- Step-wise backward selection

We require you to implement both these methods *from scratch*. You may use the Bayesian Information Criterion (BIC) to choose the subset size in each method. Do these methods eliminate one or more of the redundant predictors (if any) identified in Part (c)? In each case, fit linear regression models using the identified subset of predictors to the training set. How do the test $R^2$ scores for the fitted models compare with the model fitted in Part (b) using all predictors?




```python
# Stepwise forward selection

def get_bic(X_train, y_train):
    X_train = sm.add_constant(X_train)
    result = sm.OLS(y_train, X_train).fit()
    return result.bic

X_train = data_train.iloc[:, :-1].values
y_train = data_train.iloc[:, -1].values

X_test = data_test.iloc[:, :-1].values
y_test = data_test.iloc[:, -1].values

all_predictors = col_names[:-1]
predictors = [([], np.inf)]

for k in range(1, len(all_predictors)):
    best_k_minus_1 = predictors[-1][0]
    
    new_predictors = list(set(all_predictors) - set(best_k_minus_1))
    bics = []
    
    for predictor in new_predictors:
        
        k_predictors = best_k_minus_1 + [predictor]
        
        X_train = data_train[k_predictors].values
        
        bics.append(get_bic(X_train, y_train))
        
    best_k = best_k_minus_1 + [new_predictors[np.argmin(bics)]]
    predictors.append((best_k, np.min(bics)))

X_train = data_train[all_predictors].values
predictors.append((all_predictors, get_bic(X_train, y_train)))
```

```python
best_predictor_set = sorted(predictors, key=lambda x:x[1])[0]
best_predictor_set = ([predictor for predictor in all_predictors \
                       if predictor in best_predictor_set[0]], best_predictor_set[1])
sf_predictors = best_predictor_set[0]
X_train = data_train[best_predictor_set[0]].values
X_test = data_test[best_predictor_set[0]].values
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
result = sm.OLS(y_train, X_train).fit()
r2_test = r2_score(y_test, result.predict(X_test))

print('There are {} predictors remaining: '.format(len(best_predictor_set[0])) \
     + ', '.join(best_predictor_set[0]) + '.')
print('The test R^2 is {}.'.format(r2_test))
```

The stepwise forward selection eliminates 2 season predictors, 8 month predictors, all day of week predictors, holiday and workingday predictors, weather predictors, and a temp predictor; many redundant predictors identified in Part (c) and insignificant predicators identified in Part(b) are eliminated. However, the test $R^2$ score is slightly worse than the model fitted in Part (b) using all predictors (0.24934211146527596).

```python
# Stepwise backward selection

X_train = data_train.iloc[:, :-1].values
y_train = data_train.iloc[:, -1].values

X_test = data_test.iloc[:, :-1].values
y_test = data_test.iloc[:, -1].values

all_predictors = col_names[:-1]
predictors = [(all_predictors, get_bic(X_train, y_train))]

for k in range(len(all_predictors), 1, -1):
    best_k_predictors = predictors[-1][0]
    bics = []
    
    for predictor in best_k_predictors:
        k_minus_1 = list(set(best_k_predictors) - set([predictor]))
        X_train = data_train[k_minus_1].values
        
        bics.append(get_bic(X_train, y_train))
        
    best_k_minus_1 = list(set(best_k_predictors) - set([best_k_predictors[np.argmin(bics)]]))
    predictors.append((best_k_minus_1, np.min(bics)))
```

```python
best_predictor_set = sorted(predictors, key=lambda x:x[1])[0]
best_predictor_set = ([predictor for predictor in all_predictors \
                       if predictor in best_predictor_set[0]], best_predictor_set[1])
sb_predictors = best_predictor_set[0]
X_train = data_train[best_predictor_set[0]].values
X_test = data_test[best_predictor_set[0]].values
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
result = sm.OLS(y_train, X_train).fit()
r2_test = r2_score(y_test, result.predict(X_test))

print('There are {} predictors remaining: '.format(len(best_predictor_set[0])) \
     + ', '.join(best_predictor_set[0]) + '.')
print('The test R^2 is {}.'.format(r2_test))
```

The result of stepwise backward selection is similar to that of forward selection. The backward method keeps `temp` instead of `atemp`, and it keeps slightly different month predictors. And the test $R^2$ score is slightly worse than the model fitted in Part (b) using all predictors (0.24934211146527596).

```python
# Cross validation

def get_val_r2(X_non_test, y_non_test, train_ind, val_ind, X_test, y_test):
    X_non_test = sm.add_constant(X_non_test)
    X_test = sm.add_constant(X_test)
    X_train, X_val = X_non_test[train_ind], X_non_test[val_ind]
    y_train, y_val = y_non_test[train_ind], y_non_test[val_ind]
    result = sm.OLS(y_train, X_train).fit()
    return [r2_score(y_val, result.predict(X_val)), r2_score(y_test, result.predict(X_test))]

def cross_validation_r2(data_non_test, data_split, predictors, response='count', data_test=data_test):
    X_non_test = data_non_test[predictors].values
    y_non_test = data_non_test[response].values
    X_test = data_test[predictors].values
    y_test = data_test[response].values
    r2 = [get_val_r2(X_non_test, y_non_test, train_ind, val_ind, X_test, y_test)\
          for train_ind, val_ind in data_split]
    return r2

X_train = data_train.iloc[:, :-1].values
y_train = data_train.iloc[:, -1].values

X_test = data_test.iloc[:, :-1].values
y_test = data_test.iloc[:, -1].values

data_split = []
for train_ind, val_ind in KFold(n_splits=10).split(X_train):
    data_split.append((train_ind, val_ind))

val_r2_sf = np.array(cross_validation_r2(data_train, data_split, sf_predictors))
val_r2_sb = np.array(cross_validation_r2(data_train, data_split, sb_predictors))
val_r2_all = np.array(cross_validation_r2(data_train, data_split, all_predictors))

print('The average R^2 across all 10 validation sets for the model with the best BIC from Step-wise forward selection\
 is \n{}.'.format(np.mean(val_r2_sf, axis=0)[0]))
print('The average R^2 across all 10 validation sets for the model with the best BIC from Step-wise backward selection\
 is \n{}.'.format(np.mean(val_r2_sb, axis=0)[0]))
print('The average R^2 across all 10 validation sets for the model with all possible predictors\
 is \n{}.'.format(np.mean(val_r2_all, axis=0)[0]))
```

The result is similar for the model from stepwise forward selection and backward selection, which is better than the result using all predictors, since the model using all predictors suffers from overfitting.
